```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_path = "Images/apple.jpg"
img = cv2.imread(img_path, cv2.IMREAD_UNCHANGED)
print("shape of image:",img.shape)

img_grey = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print("image grey:",img_grey.shape)

thresh = 100
ret,thresh_img = cv2.threshold(img_grey, thresh, 255, cv2.THRESH_BINARY)
contours, hierarchy = cv2.findContours(thresh_img, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

img_contours = np.zeros(img.shape)
cv2.drawContours(img_contours, contours, -1, (255,255,255),3)
cv2.imwrite('output_data/apple_contours.jpg',img_contours) 
plt.imshow(img_contours, interpolation='none')
```

```python
def mark_object(image):
    """find the largest contour in the mask, then use it to compute the minimum enclosing circle and centroid"""
    # Finds contours in a binary image. The contours are a useful tool for shape analysis and object detection and recognition.
#     contours = cv2.con
#     contours = img_contours
# 
    c = max(contours, key=cv2.contourArea)
    ((x, y), radius) = cv2.minEnclosingCircle(c)

    # draw the circle and centroid on the frame,
    # then update the list of tracked points
    cv2.circle(image, center=(int(x), int(y)), radius=int(radius), color=(0, 255, 255), thickness=3)
    cv2.rectangle(image,(int(x) - int(radius), int(y)- int(radius)),(int(x) + int(radius), int(y)+ int(radius)),(255,0,0),3)
    return image


img_mark = mark_object(img)
img_converted = cv2.cvtColor(img_mark, cv2.COLOR_RGB2BGR)
cv2.imwrite('output_data/apple_contours_rect_cirkle.jpg', img_converted)
plt.imshow(img_mark, interpolation='none')
```
