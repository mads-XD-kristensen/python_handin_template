```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
import sklearn
from sklearn.linear_model import LinearRegression
```

```python
#downloader csven til dataframe
url = 'https://think.cs.vt.edu/corgis/datasets/csv/cars/cars.csv'
df = pd.read_csv(url)
```

```python
df.head()
```

```python
# redusere sÃ¥ dataframen kun indeholder Hondaer med fueltype gasoline
reduced_df = df.loc[df['Identification.Make']=='Honda' ]
reduced_df = reduced_df.loc[df['Fuel Information.Fuel Type']=='Gasoline']
reduced_df
```

```python
# jeg ved ikke hvad der sker her 
reduced_df = reduced_df[reduced_df['Fuel Information.Highway mpg'] < reduced_df['Fuel Information.Highway mpg'].quantile(0.90)]
```

```python
mpg = reduced_df[['Fuel Information.Highway mpg']].values
hp = reduced_df[['Engine Information.Engine Statistics.Horsepower']].values
```

```python
model = LinearRegression(normalize=True)
model.fit(hp,mpg)
print('coefficient slope',model.score(hp,mpg))
print('y er ',model.predict([[1]]), 'nÃ¥r x er 1')

plt.scatter(hp,mpg)

plt.show()
```
